# Multi-objective optimization with NSGA-II

Schaffer N1 is a 1-D 2-objective problem with an analytic Pareto front. f₁(x) = x², f₂(x) = (x-2)². Pareto-optimal x ∈ [0, 2]. We'll use NSGA-II to recover the front.

```python
import numpy as np
import matplotlib.pyplot as plt
from yee import Nsga2Config, nsga2_minimize
```

```python
def schaffer_n1(x):
    return [x[0]**2, (x[0] - 2.0)**2]
```

Run NSGA-II with pop=100, 80 generations.

```python
cfg = Nsga2Config(population_size=100, n_generations=80, seed=42)
result = nsga2_minimize(schaffer_n1, [(-2.0, 4.0)], 2, cfg)

pareto_x = result.population[result.pareto_indices, 0]
pareto_f = result.objectives[result.pareto_indices]
print(f"Pareto front size: {len(pareto_x)}")
print(f"x range in front:  [{pareto_x.min():.3f}, {pareto_x.max():.3f}]")
```

Plot the Pareto front in objective space.

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].scatter(result.objectives[:, 0], result.objectives[:, 1],
                c="lightgray", s=10, label="population")
axes[0].scatter(pareto_f[:, 0], pareto_f[:, 1],
                c="C3", s=20, label="Pareto front")
xs = np.linspace(0, 2, 200)
axes[0].plot(xs**2, (xs - 2.0)**2, "k--", alpha=0.5, label="analytic")
axes[0].set_xlabel("f1 = x²"); axes[0].set_ylabel("f2 = (x-2)²")
axes[0].legend(); axes[0].set_title("Objective space")
axes[0].grid(True, alpha=0.3)

axes[1].hist(pareto_x, bins=20, color="C3", alpha=0.7)
axes[1].axvspan(0, 2, alpha=0.2, color="gray", label="Pareto-optimal region")
axes[1].set_xlabel("x"); axes[1].set_ylabel("count")
axes[1].set_title("Decision-space spread"); axes[1].legend()
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
